```python
import tensorflow_decision_forests as tfdf
import tensorflow as tf
print(tf.__version__)
import os
import numpy as np
import pandas as pd
import tensorflow as tf
import math
```

```python
# Check the version of TensorFlow Decision Forests
print("Found TensorFlow Decision Forests v" + tfdf.__version__)
```

```python
# Load a dataset into a Pandas Dataframe.
dataset_df_train = pd.read_csv("/content/data_frame_train.csv")

# Display the first 3 examples.
dataset_df_train.head(3)
```

```python
# Load a dataset into a Pandas Dataframe.
dataset_df_test = pd.read_csv("/content/data_frame_test.csv")

# Display the first 3 examples.
dataset_df_test.head(3)
```

```python
# Name of the label column.
label = "target"
```

```python
train_ds = tfdf.keras.pd_dataframe_to_tf_dataset(dataset_df_train, label=label)
test_ds = tfdf.keras.pd_dataframe_to_tf_dataset(dataset_df_test, label=label)
```

```python
# For logging more data when training.
!pip install wurlitzer
```

```python
# Specify the model.
model_1 = tfdf.keras.RandomForestModel(verbose=2)

# Train the model.
model_1.fit(train_ds)
```

```python
model_1.compile(metrics=["accuracy"])
evaluation = model_1.evaluate(test_ds, return_dict=True)
print()

for name, value in evaluation.items():
  print(f"{name}: {value:.4f}")
```

```python
# score = model_1.predict(test_ds, verbose=2) [:2]
# print(tf.math.argmax(score[0]))

model_1.predict(test_ds, verbose=2) [:10]
```
